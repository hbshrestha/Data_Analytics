```python
def learning_rate(df):
    
    b = (math.log(df.iloc[10, 1]/df.iloc[0, 1]))/(math.log(df.iloc[10, 0]/df.iloc[0, 0]))
    
    pr = 2**b
    
    lr = 1 - pr
    
    print ("Rate of innovation: ", b)
    print("Progress ratio: ", pr)
    print ("Learning rate: ", lr)
    
    return lr
```

```python
learning_rate(solar_pv)
```

```python
learning_rate(wind_onshore)
```

```python
learning_rate(wind_offshore)
```

```python
learning_rate(csp)
```

```python
capacity = np.arange(0, 701, 50)
capacity
```

```python
def future_capital_costs(lr):
    """
    Function to get future capital costs based on learning rate.
    lr = learning rate
    #b is rate of innovation. while cost decreases with increase in production, value of b is negative.
    
    Co is $/kW 10 years ago
    Qt is  GW Cumulative installed capacity level of 2020
    Qo is GW Installed capacity around 2010
    
    """
    Co = 4731
    Qt = 710
    Qo = 40
    b = math.log(1 - lr)/math.log(2)
    
    Ct = Co * (Qt/Qo)**b
    
    return Ct
```

```python
future_capital_costs(lr = 0.4)
```

```python
df = pd.DataFrame(
                #index = capacity,
                 columns = ["LR (10%)", "LR (20%)", "LR (30%)", "LR (40%)"])

df = df.astype(float)

 
df.loc[40] = [4731, 4731, 4731, 4731]

df.loc[710] = [future_capital_costs(0.1), future_capital_costs(0.2), future_capital_costs(0.3), future_capital_costs(0.4)]


#df.interpolate(method = "linear", axis = 0, inplace = True)
```
