### 3. Localizing naive datetime object to time zone aware datetime object

The datetime index is still naive.

#### By default, the `datetime.now()` shows the local time. Therefore, you need to localize it to your specific local time zone.

```python
from datetime import datetime
from pytz import timezone

time_now = datetime.now()
print (time_now)
print (type(time_now))
print ("Time zone: ", time_now.tzinfo)
```

#### I am in Nepal right now. Therefore, I localize the datetime object to Nepali time zone.

```python
nepal_timezone = timezone("Asia/Kathmandu")
time_in_nepal = nepal_timezone.localize(time_now)
print (time_in_nepal)
print (time_in_nepal.tzinfo)
```

#### To get the current time in Germany, I need to convert time in Nepal to local time based on German time zone.

```python
german_timezone = timezone("Europe/Berlin")
time_in_germany = time_in_nepal.astimezone(german_timezone)
print (time_in_germany)
print (time_in_germany.tzinfo)
```

#### Getting current time in UTC time zone

```python
utc_timezone = timezone("UTC")
time_in_utc = time_in_nepal.astimezone(utc_timezone)
print (time_in_utc)
print (time_in_utc.tzinfo)
```

### Getting current time in Sydney time zone

```python
sydney_timezone = timezone("Australia/Sydney")
time_in_sydney = time_in_nepal.astimezone(sydney_timezone)
print (time_in_sydney)
print (time_in_sydney.tzinfo)
```

### 6. Create new dataframe for each city and convert UTC time zone to corresponding local time zone of each city

```python
df_chitwan = df[["Chitwan"]]
df_chitwan = df_chitwan.tz_convert(tz = "Asia/Kathmandu")
df_chitwan
```

```python
df_newyork = df[["NewYork"]]
#EST5EDT. US Eastern Time. 5 hours west of UTC
df_newyork = df_newyork.tz_convert(tz = "EST5EDT")
df_newyork.head()
```

```python
df_bonn = df[["Bonn"]]
df_bonn = df_bonn.tz_convert(tz = "Europe/Berlin")
df_bonn.head()
```

```python
df_sydney = df[["Sydney"]]
df_sydney = df_sydney.tz_convert(tz = "Australia/Sydney")
df_sydney.head()
```

### 7. Comparing the plots of solar irradiance data between UTC time zone and local time zone of each city

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (20, 6))

fig.suptitle("Solar irradiance on October 1, 2020")
ax1.plot(df.loc["2020-10-01"])
ax1.set_title("Based on UTC time zone")
ax1.xaxis.set_ticks(ticks = df.loc["2020-10-01"].index[::4], labels = np.arange(0, 24, 4))

cities = df.columns.tolist()
handles = ax1.get_legend_handles_labels()[0]
ax1.legend(handles, labels = cities, loc = "upper right")

ax1.set_xlabel("Hour of day")
ax1.set_ylabel("W/m$^2$")


ax2.plot(df_chitwan.loc["2020-10-01"].values.tolist())
ax2.plot(df_newyork.loc["2020-10-01"].values.tolist())
ax2.plot(df_bonn.loc["2020-10-01"].values.tolist())
ax2.plot(df_sydney.loc["2020-10-01"].values.tolist())
ax2.xaxis.set_ticks(ticks = np.arange(0, 24, 4), labels = np.arange(0, 24, 4))

handles = ax2.get_legend_handles_labels()[0]
ax2.legend(handles, labels = cities)

ax2.set_title("Based on local time zone of each city/country")
ax2.set_xlabel("Hour of day")
ax2.set_ylabel("W/m$^2$")

plt.savefig("")

plt.show()
```
